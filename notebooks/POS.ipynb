```python
import spacy
npl = spacy.load('en_core_web_sm')

import docx

def read_doc(file):
    doc = docx.Document(file)
    fullText = []
    for para in doc.paragraphs:
        fullText.append(para.text)
    return '\n'.join(fullText)


text = read_doc('C:\\dev\\nlp_practice\\data\\Brexit.docx')


# normalize whitespace
text = text.replace('\xa0', ' ')

doc = npl(text)
text
```

```python
import pandas as pd
pos_data = []
for token in doc:
    pos_data.append({
        "Token": token.text,
        "POS": token.pos_,
        "Tag": token.tag_,
        "Dep": token.dep_,
        "Shape": token.shape_,
        "Is Alpha": token.is_alpha,
        "Is Stop": token.is_stop
    })

pos_df = pd.DataFrame(pos_data)
pos_df
```

```python
ner_data = []
for ent in doc.ents:
    ner_data.append({
        "Text": ent.text,
        "Start Char": ent.start_char,
        "End Char": ent.end_char,
        "Label": ent.label_
    })

ner_df = pd.DataFrame(ner_data)
ner_df
```
